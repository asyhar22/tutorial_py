##### Import library pandas

```python
import pandas as pd
```

##### Simpan tabel html ke dalam variabel `tabel_html`

```python
tabel_html = '''<table class="table table-striped table-no-bordered table-lg">
        <thead>
            <tr>
                <th scope="col" class="text-center">
                    Tanggal</th>
                <th scope="col" class="text-center">
                    Data Inflasi
                </th>
            </tr>
        </thead>


                <tbody>

                <tr>
                    <td class="text-center">Juni 2023</td>
                    <td class="text-center">3.52 %</td>
                </tr>

                <tr>
                    <td class="text-center">Mei 2023</td>
                    <td class="text-center">4 %</td>
                </tr>

                <tr>
                    <td class="text-center">April 2023</td>
                    <td class="text-center">4.33 %</td>
                </tr>

                <tr>
                    <td class="text-center">Maret 2023</td>
                    <td class="text-center">4.97 %</td>
                </tr>

                <tr>
                    <td class="text-center">Februari 2023</td>
                    <td class="text-center">5.47 %</td>
                </tr>

                <tr>
                    <td class="text-center">Januari 2023</td>
                    <td class="text-center">5.28 %</td>
                </tr>

                <tr>
                    <td class="text-center">Desember 2022</td>
                    <td class="text-center">5.51 %</td>
                </tr>

                <tr>
                    <td class="text-center">November 2022</td>
                    <td class="text-center">5.42 %</td>
                </tr>

                <tr>
                    <td class="text-center">Oktober 2022</td>
                    <td class="text-center">5.71 %</td>
                </tr>

                <tr>
                    <td class="text-center">September 2022</td>
                    <td class="text-center">5.95 %</td>
                </tr>

                </tbody>

    </table>'''
```

##### Baca tabel HTML tersebut menggunakan metode `pandas.read_html` sehingga menghasilkan sebuah list

```python
my_table = pd.read_html(tabel_html)
my_table
```

##### Ubah list tersebut ke format `DataFrame`

```python
df = my_table[0]
df
```

##### Simpan data frame menjadi file excel

```python
# simpan ke excel
df.to_excel('tabel_inflasi.xlsx', index=False)
```
